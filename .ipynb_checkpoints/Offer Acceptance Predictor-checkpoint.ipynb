# Offer acceptance predictor model
### Or, _"My First Neural Network"_
In this notebook, we will:
* Generate some random job offer data for graduate jobs
* Create a toy model that gives the probability a candidate might accept an offer given the input data
* Randomly accept/reject offers based on that probability
* Build and train a neural network that can predict whether a candidate accepts or rejects (i.e. to try and replicate the toy model)
* Use the model to tweak a company's offer to increase the chance the candidate will accept

```python
# Import modules
import tensorflow as tf
#from tensorflow import keras
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
```

## Generate data:
In the absence of a real dataset, make some assumptions about what real life data looks like. Inputs will be:
* The salary included in the offer
* The days of annual leave included in the offer
* The number of employees in the company
* The turnover of the company
* The company's remote working policy (e.g. fully office/remote or hybrid)
Make sure days leave and number of employees are generated as integers - no fractions of a day of leave or half employees!

```python
# Randomly generate some data
n = int(1e6)

# Assume offers are normally distributed about £30k with a width of £1k
# Assert no offers are less than £15k or more than £50k
lower, upper = 1.5e4, 5e4
mu, sigma = 3.0e4, 1e4
offer = stats.truncnorm((lower - mu) / sigma,
                        (upper - mu) / sigma,
                        loc=mu, scale=sigma).rvs(n)
# Assume holidays are normally distributed about 25 days, with a width of 2 days, ignoring bank holidays
# Assert no offers include less than the statutory leave allownace or more than 30 days EXCEPT FOR
# offers of 31 days are actually from companies with infinite holiday policies
lower, upper = 20, 32
mu, sigma = 25, 2
holiday = stats.truncnorm((lower - mu) / sigma,
                          (upper - mu) / sigma,
                          loc=mu, scale=sigma).rvs(n).astype(int)
holiday = holiday.astype(float)
holiday[holiday == 31] = np.inf     # where 

# Assume the logarithm of the number of employees in a company is normally distributed about 2.5 with a width of 1
# Assert there are no companies with less than 1 employee or more than 100,000
lower, upper = 0, 5,
mu, sigma = 2.5, 1
employees = (10**stats.truncnorm((lower - mu) / sigma,
                                 (upper - mu) / sigma,
                                 loc=mu, scale=sigma).rvs(n)).astype(int)

# Assume a quarter of the offers include hybrid working
p = 1/4
hybrid = np.random.choice(a=[True, False], size=(n), p=[p, 1-p])

# Assume a quarter of those offers including hybrid working allow fully remote working
p = 1/4
remote = np.random.choice(a=[True, False], size=(n), p=[p, 1-p]) & hybrid

# Assume turnover is related to the square root of the number of employees, give or take
turnover = 1e4 * employees**0.5 * 10**np.random.normal(1, 0.2, n)
```

# Create a toy model
Create functions that normalise the data between 0 and 1 so that probabilities can be generated by weighting the inputs.
Assume:
* the probability of offer acceptance is related to the order of magnitude of the offer (significance of the difference between £15k and £17k is larger than the significance of the difference between £48k and £50k)
* the significance of the number of holidays exponentially decays, such that statutory holidays returns 0 and infinite holidays returns 1
* candidates like to work for small start-up companies because they're exciting, but candidates also like to work for large companies because they are stable and have "big names" (e.g. Microsoft, Google)
* the probability of the offer acceptance is related to the order of magnitude of turnover

```python
# Normalise data
def normalise(A, a=None):
    if a is None: a = A
    return (a - np.min(A))/np.ptp(A)

offer_n = normalise(np.log(offer))
holiday_n = 1 - np.exp(1 - holiday/20)
employees_n = normalise(1/(1+np.log(employees)) + np.log(employees)/10)
turnover_n = normalise(np.log(turnover))
hybrid_n = hybrid.astype(int)
remote_n = remote.astype(int)
```

Create a function to combine all the normalised inputs based on weights. Assume beyond a certain point, offers are too good to refuse - therefore don't divide by the sum of the weights. Make sure no probabilities are greater than 1.

```python
# Function for probability of acceptance
def acceptance_probability(variables, weights, factor=1):
    p = (variables.T * weights).sum(axis=1) * factor
    p[p>1] = 1
    return p

# Generate acceptance probabilities
variables =           np.array([offer_n, holiday_n, employees_n, turnover_n, hybrid_n, remote_n])
weights   = normalise(np.array([9,       6,         4,           2,          3,        4])) / len(variables)
factor = 5
acceptance = acceptance_probability(variables, weights, factor)
```

Demonstrate the spread of the toy model by plotting the probability of acceptance against the offer. Note the probability is capped at 1.

```python
fig = plt.figure()
ax = fig.add_subplot()
hb = ax.hexbin(offer/1e3, acceptance, gridsize=128, mincnt=1, bins='log')
ax.set_xlabel('Offer (£1000s)')
ax.set_ylabel('Probability of acceptance')
cb = fig.colorbar(hb, ax=ax)
cb.set_label('Counts')
```

Randomly accept or reject offers based on their probability.

```python
# Decide if accepted or rejected
accepted = np.where(np.random.uniform(size=n) < acceptance, True, False)
print('Fraction of all offers accepted: {0:.2f}'.format(accepted.mean()))
```

# Build and train a neural network to replicate the toy model

```python
#from tensorflow.keras import datasets, layers, models

X = variables.T
y = accepted.astype(int)

# Build
model = tf.keras.Sequential([tf.keras.layers.Dense(12, input_shape=(X.shape[1],), activation='relu'),
                             tf.keras.layers.Dense(9, activation='relu'),
                             tf.keras.layers.Dense(4, activation='relu'),
                             tf.keras.layers.Dense(1, activation='sigmoid'), # The last layer must output a 0 or 1, hence sigmoid activation
                             ])

# Compile
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])

#train
history = model.fit(X, y, verbose = 2,
                    validation_split = 0.5,
                    epochs=10, 
                    batch_size = 1024)
```

Plot the accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy','validation accuracy'])
```

# Use the model to optimise a company's offer.
**First, show how the number of employees and remote working policy affects the probability of acceptance of an offer of £25k and statutory leave allowances. Assume turnover is the average turnover for the given company size.**

```python
i = 101

# Create some data and normalise in the same way as the inputs to the model
o = np.array([25000]*i)
h = np.array([20]*i)

o_n = normalise(offer, o)
h_n = 1 - np.exp(1 - h/20)

e = np.logspace(0,5,i)
t = 1e4 * e**0.5
e_n = normalise(1/(1+np.log(employees)) + np.log(employees)/10, 1/(1+np.log(e)) + np.log(e)/10)
t_n = normalise(np.log(turnover), np.log(t))

# Use the model to predict the acceptance based on the inputs
office = model.predict(np.array([o_n, h_n, e_n, t_n, np.array([0]*i), np.array([0]*i)]).T)
hybrid = model.predict(np.array([o_n, h_n, e_n, t_n, np.array([1]*i), np.array([0]*i)]).T)
remote = model.predict(np.array([o_n, h_n, e_n, t_n, np.array([1]*i), np.array([1]*i)]).T)

# Plot the results
fig = plt.figure()
ax = fig.add_subplot()
ax.semilogx(e, office, label='Office')
ax.semilogx(e, hybrid, label='Hybrid')
ax.semilogx(e, remote, label='Remote')
ax.set_ylim([0, 1.05])
ax.set_yticks(np.linspace(0, 1, 11))
ax.set_xlabel('Employees')
ax.set_ylabel('Probability offer accepted')
ax.grid()
ax.legend()
```

**Show how a company with a given number of employees can change the probability by changing the offer amount.**

```python
i = 101

h = np.array([20]*i)
e = np.array([1000]*i)
t = 1e4 * e**0.5
h_n = 1 - np.exp(1 - h/20)
e_n = normalise(1/(1+np.log(employees)) + np.log(employees)/10, 1/(1+np.log(e)) + np.log(e)/10)
t_n = normalise(np.log(turnover), np.log(t))

o = np.linspace(15000,45000,i)
o_n = normalise(offer, o)

# Use the model to predict the acceptance based on the inputs
office = model.predict(np.array([o_n, h_n, e_n, t_n, np.array([0]*i), np.array([0]*i)]).T)
hybrid = model.predict(np.array([o_n, h_n, e_n, t_n, np.array([1]*i), np.array([0]*i)]).T)
remote = model.predict(np.array([o_n, h_n, e_n, t_n, np.array([1]*i), np.array([1]*i)]).T)

# Plot the results
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(o/1e3, office, label='Office')
ax.plot(o/1e3, hybrid, label='Hybrid')
ax.plot(o/1e3, remote, label='Remote')
ax.set_ylim([0, 1.05])
ax.set_yticks(np.linspace(0, 1, 11))
ax.set_xlabel('Offer (£1000s)')
ax.set_ylabel('Probability offer accepted')
ax.grid()
ax.legend()
```
